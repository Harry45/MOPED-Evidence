```python
import numpy as np 
import matplotlib.pylab as plt 

import utils.helpers as hp

plt.rc('text', usetex=True)
plt.rc('font',**{'family':'sans-serif','serif':['Palatino']})
figSize  = (12, 8)
fontSize = 20
```

# Nested Models


The data is being generated using the two-parameter model, that is, $ax^{2}+bx$, with $\sigma=0.2$ and $N=20$. The fiducial point is $\boldsymbol{\theta}=[-1,4]$. We are then comparing two models, $y=ax^{2}+bx$ $(\mathcal{M}_{2})$ and $y=ax^{2}+bx+c$ $(\mathcal{M}_{3})$.

$$
B_{23} = \dfrac{Z_{2}}{Z_{3}}
$$

```python
nested = hp.load_arrays('results', 'nested')
nested_moped = hp.load_arrays('results', 'nested-moped')
```

```python
plt.figure(figsize = (10,10))
plt.hist(nested, density = True, label = 'No Compression', bins = 20, alpha = 0.5)
plt.hist(nested_moped, density = True, label = 'Compression', bins = 20, alpha = 0.5)
# plt.axvline(0, linestyle = '--')
# plt.xlim(0, 6)
plt.xlabel(r'$\textrm{log } B_{23}$', fontsize = fontSize)
plt.tick_params(axis='x', labelsize=fontSize)
plt.tick_params(axis='y', labelsize=fontSize)
plt.legend(loc = 'best',prop={'family':'sans-serif', 'size':fontSize})
plt.show()
```

# Non-Nested 

The data is generated using the sinusoidal model:

$$
y = A\,\textrm{sin}\,\omega x
$$

and the other model is 

$$
y = ax^{2} + bx
$$

The Bayes factor is:


$$
B_{\textrm{sin}|\textrm{quad}} = \dfrac{Z_{\textrm{sin}}}{Z_{\textrm{quad}}}
$$

```python
nonnested = hp.load_arrays('results', 'non-nested')
nonnested_moped = hp.load_arrays('results', 'non-nested-moped')
```

```python
plt.figure(figsize = (10,10))
plt.hist(nonnested, density = True, label = 'No Compression', bins = 20, alpha = 0.5)
plt.hist(nonnested_moped, density = True, label = 'Compression', bins = 20, alpha = 0.5)
# plt.axvline(0, linestyle = '--')
# plt.xlim(0, 6)
plt.xlabel(r'$\textrm{log } B_{23}$', fontsize = fontSize)
plt.tick_params(axis='x', labelsize=fontSize)
plt.tick_params(axis='y', labelsize=fontSize)
plt.legend(loc = 'best',prop={'family':'sans-serif', 'size':fontSize})
plt.show()
```
