```python
import numpy as np 
import matplotlib.pylab as plt 

import utils.helpers as hp

plt.rc('text', usetex=True)
plt.rc('font',**{'family':'sans-serif','serif':['Palatino']})
figSize  = (12, 8)
fontSize = 20
```

# Nested Models

```python
results = hp.load_pickle('results', 'nested')
```

```python
nrealisations = len(results)
evi_uncomp = [results[i]['bf_12'] for i in range(nrealisations)]
evi_comp = [results[i]['bf_moped_12'] for i in range(nrealisations)]
```

```python
plt.figure(figsize = (10,10))
plt.hist(evi_uncomp, density = True, label = 'No Compression', bins = 20)
plt.hist(evi_comp, density = True, label = 'Compression', bins = 20)
# plt.axvline(0, linestyle = '--')
plt.xlim(0, 6)
plt.xlabel(r'$\textrm{log } B_{12}$', fontsize = fontSize)
plt.tick_params(axis='x', labelsize=fontSize)
plt.tick_params(axis='y', labelsize=fontSize)
plt.legend(loc = 'best',prop={'family':'sans-serif', 'size':fontSize})
plt.show()
```
